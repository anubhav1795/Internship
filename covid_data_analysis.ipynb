```python
# Import the Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#delete the usless columns
df = dataset.drop (["Lat","Long"],axis = 1,inplace =True)
```

```python
dataset.head()
```

```python
#aggregate the rows by the country 
corona_dataset_aggregated = dataset.groupby("Country/Region").sum(numeric_only=True)
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#visualize data related to countries
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
#calculate a good measure 
corona_dataset_aggregated.loc["China"].plot()
```

```python
#To check the spread of corona_virus for the first three days in china
corona_dataset_aggregated.loc["China"][:3].plot()
```

observation: First day 550 cases,second day 650 cases and from third day it crossed more than 900 cases per day.

```python
# calculate the first derivative of the curve  for the rate of infection.
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
 #maximum infection rate 
```

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
countries =list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rates"] =max_infection_rates  
```

```python
corona_dataset_aggregated
```

```python
# crate a dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```

```python
corona_data
```

```python
#importing the dataset 2
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report
```

```python
#dropping the useless column
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.drop(useless_cols, axis=1, inplace=True)
happiness_report.head()
```

```python
happiness_report.set_index("Country or region",inplace = True)
happiness_report.head()
```

```python
#join the dataset
corona_data.shape
```

```python
happiness_report.shape
```

```python
data = corona_data.join(happiness_report, how ="inner")
```

```python
data
```

```python
data.corr()
```

#observation 

GDP and Social support have positive correlation.

There is very low correlation between the Max_infection_rates and Freedom to makelife choices.

High correlation b/w Healthy life expectancy and Gdp per capita
